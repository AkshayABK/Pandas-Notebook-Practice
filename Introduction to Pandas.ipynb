```python
import pandas as pd
```

## Ways To Create The Data Frame

1. Pandas DataFrame with Two-dimensional List

```python
ls_two_dimension = [[1,'Akshay'], [2,'Amarnath'], [3, 'Ganesh']]

df_two_dimension = pd.DataFrame(ls_two_dimension, columns=['id', 'Name'])

df_two_dimension
```

2. Pandas DataFrame from dict of narray/lists

```python
dic_to_ls = {'id':[1,2,3], 'Name':['Akshay', 'Amarnath', 'Adarsh']}

df_dic_to_ls = pd.DataFrame(dic_to_ls)

df_dic_to_ls
```

3. Pandas dataframe using list of tuples

```python
ls_of_tuple = [(1,'Akshay'),(2,'Amarnath'),(3,'Adarsh'), (4, 'Ganesh'), (5, 'Alok')]

df_ls_of_tuple = pd.DataFrame(ls_of_tuple, columns=['id', 'Name'])

df_ls_of_tuple
```

4. Pandas DataFrame from List of Dicts

```python
ls_of_dic = [{'id':1, 'name':'Akshay'},{'id':2, 'name':'Amarnath'}]
                                                          
df_ls_of_dic = pd.DataFrame(ls_of_dic)

df_ls_of_dic
```

5. Convert list of nested dictionary into Pandas dataframe

```python
countries = {
    "1": {"Country": "New Country 1",
        "Capital": "New Capital 1",
        "Population": "123,456,789"},
    "2": {"Country": "New Country 2",
        "Capital": "New Capital 2",
        "Population": "987,654,321"},
    "3": {"Country": "New Country 3",
        "Capital": "New Capital 3",
        "Population": "111,222,333"}
}

df_nested_list = pd.DataFrame.from_dict(countries, orient='index')

df_nested_list
```

## Creating Genaral Data Frame 

```python
df_general_ls = [[1,'Akshay',3000], [2,'Amarnath',5000], [3, 'Ganesh',6000], [4, 'Adarsh',9000], [5,'Akshay',7000]]

df_general = pd.DataFrame(df_general_ls, columns=['ID', 'Name', 'Salary'])

df_general
```

## Returning Perticular Column

```python
df_general[['Name']]
```

## Filtering the Rows based on Condition

1. Filtering the recording and returning entire DF

```python
df_filter_01 = df_general[df_general.Salary <= 5000]

df_filter_01
```

2. Filtering the record and returning the required column

```python
df_filter_02 = df_general[df_general.Salary <= 5000][['Name', 'Salary']]

df_filter_02
```

## Adding New Column to Existed DF

```python
df_general['Bonus'] = df_general['Salary']*2

df_general
```

## Adding New Row To Existing DF

1. Concat

Note: Running this query twice will result in the duplication of the same record.

```python
new_row = [[6, 'Alok', 1000,'6000']]

df_general = pd.concat([df_general, pd.DataFrame(new_row, columns=['ID', 'Name', 'Salary', 'Bonus'])], ignore_index=True)

new_row = [[6, 'Alok', 1000,'6000']]

df_general = pd.concat([df_general, pd.DataFrame(new_row, columns=['ID', 'Name', 'Salary', 'Bonus'])], ignore_index=True)

new_row = [[6, 'Alok',None,'6000']]

df_general = pd.concat([df_general, pd.DataFrame(new_row, columns=['ID', 'Name', 'Salary', 'Bonus'])], ignore_index=True)

df_general
```

## Pivot 

```python
Pivot_data = {
    'City': ['Jacksonville', 'Jacksonville', 'Jacksonville', 
             'Jacksonville', 'Jacksonville', 'ElPaso', 'ElPaso', 
             'ElPaso', 'ElPaso', 'ElPaso'
            ],
    
    'Month': ['January', 'February', 'March', 'April', 'May', 
              'January', 'February', 'March', 'April', 'May'
             ],
    
    'Temperature': [13, 23, 38, 5, 34, 20, 6, 26, 2, 43]
}

pivot_df = pd.DataFrame(Pivot_data)

pivot_df
```

## Method Chaining

- Problem Statement
    - Write a solution to list the names of animals that weigh strictly more than 100 kilograms.
    - Return the animals sorted by weight in descending order.

```python
prac_data = {
    'Name': ['Tatiana', 'Khaled', 'Alex', 'Jonathan', 'Stefan', 'Tommy'],
    'Species': ['Snake', 'Giraffe', 'Leopard', 'Monkey', 'Bear', 'Panda'],
    'Age': [98, 50, 6, 45, 100, 26],
    'Weight': [464, 41, 328, 463, 50, 349]
}

prac_df = pd.DataFrame(prac_data)

prac_df
```

```python
filtered_df = prac_df[prac_df['Weight']>=100][['Name']]

filtered_df
```

```python
filtered_df.sort_values(by=['Name'], ascending = False)
```

```python
prac_df[prac_df['Weight']>=100].sort_values(by='Name', ascending = False)[['Name']]
```
